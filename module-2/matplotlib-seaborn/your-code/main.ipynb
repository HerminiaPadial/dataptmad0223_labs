```python
#Agrego una nueva columna al DF cuyos valores son la media de pasos por mes y wewk_or_weekend
#Hay un error en el nombre de la columna week_or weekend
#transform(aplica funciones sobre los grupos generados por groupby y agregando la información resultante al DF original)
#Aggregate, filter y apply (generan un dataframe diferente)
fitbit['average_steps_month'] = fitbit.groupby(['Months_encoded',fitbit.columns[17]])['Steps'].transform('mean')
```
